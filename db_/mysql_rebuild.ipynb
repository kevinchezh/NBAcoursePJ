## Import data
The following cells import reg and post season player and team data into pandas.
We will insert data further by using the pandas dataframe.

```python
#we import the json doc and sparse them into pandas dataframe. The team jsons are a little different from player jsons. But most steps are the same.
flag = False
post_team_pd = None
for i in range(2000,2018):
    filename = str(i)+'post_team.json'
    with open(filename,'r') as f:
        data = json.load(f)
#     print(i)
    column = data['resultSets'][0]['headers']
    data_pd = pd.DataFrame(columns = column)
    data_sets = data['resultSets'][0]['rowSet']
    for j in range(len(data_sets)):
        data_pd.loc[j] = data_sets[j]
    data_pd['year'] = i
    if not flag:
        post_team_pd = data_pd
        flag = True
    else:
           post_team_pd = post_team_pd.append(data_pd)
    post_team_pd.reset_index()
post_team_pd.info()
```

```python
flag = False
reg_team_pd = None
for i in range(2000,2018):
    filename = str(i)+'reg_team.json'
    with open(filename,'r') as f:
        data = json.load(f)
#     print(i)
    column = data['resultSets'][0]['headers']
    data_pd = pd.DataFrame(columns = column)
    data_sets = data['resultSets'][0]['rowSet']
    for j in range(len(data_sets)):
        data_pd.loc[j] = data_sets[j]
    data_pd['year'] = i
    if not flag:
        reg_team_pd = data_pd
        flag = True
    else:
           reg_team_pd = reg_team_pd.append(data_pd)
reg_team_pd.reset_index()
```

```python
# data cleaning and wrangling. After we first build reg_team, we find some of the teams have two names. That' s why we rename them so we will get a 
# clean dataframe.
for i in range(reg_team_pd.shape[0]):
    if reg_team_pd.iloc[i]['year'] == 2015 or reg_team_pd.iloc[i]['year']==2016 or reg_team_pd.iloc[i]['year']==2017:
        if reg_team_pd.iloc[i]['TEAM_NAME'] == 'LA Clippers':
            reg_team_pd.iloc[i]['TEAM_NAME'] = 'Los Angeles Clippers'
reg_team_pd[reg_team_pd.TEAM_NAME == 'LA Clippers']
```
